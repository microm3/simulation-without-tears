# Transformation step verification
Runs the two pipeline steps in order:
1) transform catalog OntoUML JSON to Alloy files
2) verify generated Alloy models for parse + executability (batch run).

To run with a different version of OntoUML2Alloy, compile a new version of the library and include in ```src/main/resource/ontouml2alloy```. The pre-update version is included in a separate folder in ```src/main/resource/```, to use it, manually rename the folder to ```ontouml2alloy```

````python
import pathlib

version_path = pathlib.Path("../../src/main/resources/ontouml2alloy/VERSION")
print("OntoUML2Alloy version (git commit hash):", version_path.read_text().strip())
````

## 1) Transform catalog JSON to Alloy batch output

Uses the crawl manifest as input and writes generated Alloy files to `evaluation/transformation-verification/results-baseline/batch-alloy-output`.

````python
%%bash
set -e

cd ../../

OUT_DIR="evaluation/transformation-verification/results-baseline/batch-alloy-output"
EXCLUDE_SLUGS="mgic-antt2011"
# EXCLUDE_SLUGS=""

rm -rf "$OUT_DIR"


mvn -q clean compile exec:java \
  -Dexec.mainClass="com.example.batch.CatalogTransformationRunner" \
  -Dexec.args="evaluation/catalog-download/ontouml_downloads/manifest.json $OUT_DIR ${EXCLUDE_SLUGS} --no-metadata"
````

## 2) Verify generated Alloy models (parse + executability)

Reads the transform manifest from `evaluation/transformation-verification/results-baseline/batch-alloy-output` and writes verification JSON to `evaluation/transformation-verification/results-baseline/verification`.

Optional **slug** exclusions: set `EXCLUDE_SLUGS_CSV` in the next cell (comma-separated catalog slugs; slugs do not contain commas).

````python
%%bash
set -e

export MAVEN_OPTS='-Xmx8g'

# Optional comma-separated catalog slugs to skip
# EXCLUDE_SLUGS_CSV="barcelos2015transport-networks,mgic-antt2011"
EXCLUDE_SLUGS_CSV=""


cd ../../

mvn clean -q compile exec:java \
  -Dexec.mainClass="com.example.batch.CatalogAlloyRunner" \
  -Dexec.args="evaluation/transformation-verification/results-baseline/batch-alloy-output/alloy_batch_manifest.json evaluation/transformation-verification/results-baseline/verification/verification_results.json ${EXCLUDE_SLUGS_CSV}"
````
