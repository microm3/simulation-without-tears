# Catalog Download

Scripts to download all OntoUML models from the [OntoUML/UFO Catalog](https://scs-ontouml.eemcs.utwente.nl) catalog

## Download model index
Download an updated catalog model index from the OntoUML API as Turtle RDF into `curl_ontouml_catalog.ttl`, the indexed models will be downloaded later. 

```python
%%bash
set -e
curl -sS "https://scs-ontouml.eemcs.utwente.nl/catalog/b663ca18-8085-44a7-bcfe-2c2b5ba1faa8/page/model?page=0&size=2000" -o curl_ontouml_catalog.ttl
```

## Run download script

Reads `curl_ontouml_catalog.ttl` and downloads the json-distribution.

Output: `ontouml_downloads/` with `manifest.json`.

```python
%%bash
set -e

kotlin catalog-download.main.kts curl_ontouml_catalog.ttl ontouml_downloads
```
